```python
x = np.linspace(0, 10, 5)
y1 = x**2
y2 = x**3
y3 = x**0.5
y4 = np.sin(x**2)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5)) #создание Figure() и Axes()

ax[0, 0].plot(x, y1, 'r') #назначение ячеек графикам
ax[0, 1].plot(x, y2, 'b')
ax[1, 0].plot(x, y3, 'g')
ax[1, 1].plot(x, y4, 'c')

ax[0, 0].grid() #подруб сетки
ax[0, 1].grid()
ax[1, 0].grid()
ax[1, 1].grid()

plt.show()
```

```python
x = np.linspace(0, 10, 5)
y1 = x**2
y2 = x**3
y3 = x**0.5
y4 = np.sin(x**2)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 5))
  #альтернативное задание картежем с присвоением графикам имени

ax1.plot(x, y1, 'r') #построение
ax2.plot(x, y2, 'b')
ax3.plot(x, y3, 'g')
ax4.plot(x, y4, 'c')

ax1.set_title('y1') #установка заголовков
ax2.set_title('y2')
ax3.set_title('y3')
ax4.set_title('y4')

fig.suptitle('Графики') # общий супер-заголовок

plt.tight_layout()
plt.show()
```

```python
from matplotlib.gridspec import GridSpec

x = np.linspace(-np.pi, np.pi, 30)
y1 = np.sin(x)
y2 = 0.5 * np.sin(x)
y3 = 2 * np.sin(x)

fig = plt.figure(figsize=(10, 5))

widths = [1, 2, 2]
heights = [2, 2, 1]
#задаем масштаб ячеек по ширине и высоте

gs = GridSpec(nrows=3, ncols=3, figure=fig, width_ratios=widths, height_ratios=heights)
#создаем сетку 3х3, передаем соотношения ширины и высоты

ax1 = fig.add_subplot(gs[0,0]) #поочередно наносим графики
ax1.plot(x, y1, 'y')
ax2 = fig.add_subplot(gs[0,1])
ax2.plot(x, y2, 'g')
ax3 = fig.add_subplot(gs[1:,:])
ax3.plot(x, y3, 'm')

plt.tight_layout()
plt.show()

fig.savefig('my_plot.jpg', bbox_inches='tight', dpi=200) #исправление ошибки
                                                         #при отображении осей
                                                         #при сохранении
```

### **ScatterPlot**

```python
x = 'petal length (cm)'
y = 'petal width (cm)'

groups = df.groupby('target') #группировка по значениям столбца на три группы

for name, group in groups: #name - значение ключа группы, group - сама группа
    plt.scatter(group[x], group[y], label=name, edgecolor='black', linewidth=0.3, alpha=0.6)

plt.legend()
plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

plt.show()
```

```python
x = df['petal length (cm)']
y = df['petal width (cm)']
z = df['sepal width (cm)'] / df['sepal length (cm)']

plt.scatter(x, y, c=z, cmap='PuOr_r') #цветовая палитра и принцип раскраски
plt.xlabel('Длина лепестка')
plt.ylabel('Ширина лепестка')

cbar = plt.colorbar() #столбик справа
cbar.set_label('Отношение ширины чашелистика к его длине')

plt.show()
```

### **BarPlot, BarhPlot**

```python
df = pd.DataFrame(df_values, columns=df_columns)

plt.figure(figsize=(8, 6))

ax = plt.bar(df['Год'], df['Торговые площади X5Group, тыс. кв.м.'], 
        width=0.4, color='green', alpha=0.6)

plt.title('Торговые площади X5Group')
plt.xlabel('Год')
plt.ylabel('Торговые площади, тыс. кв.м.')

plt.bar_label(ax, padding=5) #подпись к столбцам и ее удаление от столбца

plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.barh(df['Год'], df['Торговые площади Магнит, тыс. кв.м.'], 
        height=0.5, color='red', alpha=0.5)

plt.title('Торговые площади ПАО "Магнит"')
plt.xlabel('Год')
plt.ylabel('Торговые площади, тыс. кв.м.')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

w = 0.1 #настройка сдвига для предотвращения наложения столбцов
x = df.index.values

plt.bar(x-w, df['Торговые площади Магнит, тыс. кв.м.'],
        width=w*2, color='red', alpha=0.5)

plt.bar(x+w, df['Торговые площади X5Group, тыс. кв.м.'], 
        width=w*2, color='green', alpha=0.5)

plt.xticks(ticks=x, labels=df['Год']) #изменение подписей на оси x

plt.legend(['Магнит', 'X5Group'])
plt.title('Торговые площади')
plt.xlabel('Год')
plt.ylabel('Торговые площади, тыс. кв.м.')

plt.show()
```

### **StackPlot**

```python
y1 = df['Цех №1']
y2 = df['Цех №2']
y3 = df['Цех №3']
x = df['День недели']
labels = ['Цех №1', 'Цех №2', 'Цех №3']

plt.figure(figsize=(8, 5))

plt.stackplot(x, y1, y2, y3, 
              labels=labels)

plt.ylim(0, 1000)
plt.grid(alpha=0.3)
plt.legend()

plt.show()
```

```python
y1 = df['Цех №1']
y2 = df['Цех №2']
y3 = df['Цех №3']
x = df['День недели']
labels = ['Цех №1', 'Цех №2', 'Цех №3']
colors = ['#05B893', '#F59FC6', '#D7F59F']

plt.figure(figsize=(8, 5))

plt.stackplot(x, y1, y2, y3, 
              labels=labels,
              colors=colors,
              hatch='///')

plt.ylim(0, 1000)
plt.grid(alpha=0.3)
plt.legend()
plt.rcParams['hatch.color'] = 'lightgrey'

plt.show()
```
